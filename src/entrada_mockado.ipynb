# Gerando o Dataset De Entrada Mockado

### Importando bibliotecas

```python
import numpy as np
import pandas as pd
```

### Lendo o Arquivo de Entrada de Acidentes
(Saida do notebook anterior)

```python
DatasetEntradaAcidentes = pd.read_csv(
    "../dados/modificados/acidentes_entrada.csv"
)
DatasetEntradaAcidentes.shape
```

### Adicionando coluna HH OS Mockada para o Modelo
#### Gerando floats aleatorios no intervalo:
0 a 1350 (valor maximo de HH somado das OS por mes)

```python
DatasetEntradaAcidentes['HH OS (Mockado Uniform)'] = np.random.uniform(
    0, 1350, len(DatasetEntradaAcidentes))

DatasetEntradaHHmockado = DatasetEntradaAcidentes
DatasetEntradaHHmockado
```

### Moldando distribuição do HH OS mockado:
#### Aproximando da realidade dos dados

```python
DatasetEntradaHHmockado = DatasetEntradaAcidentes

tamanho_amostra = len(DatasetEntradaHHmockado)

# Gerando valores seguindo uma distribuição exponencial
# scale ajusta a inclinação da distribuição
valores_exponenciais = np.random.exponential(scale=200, size=tamanho_amostra)

# Escalando os valores para o intervalo desejado, garantindo que não ultrapassem 1350
valores_escalados = np.minimum(valores_exponenciais, 1350)

# Adicionando a nova coluna ao DataFrame
DatasetEntradaHHmockado['HH OS (Mockado Exponential)'] = valores_escalados

DatasetEntradaHHmockado
```

### Transformando o Dataset de Entrada HH OS Mockado em CSV

```python
DatasetEntradaHHmockado.to_csv(
    "../dados/modificados/entrada_modelo_HH_OS_mockado.csv", index=False
)
```
