```python
import numpy as np
import pandas as pd
from resolve_path import ajuste_path, read_input
```

```python
path = "data/input/"
df = read_input(path + "IW49_OP_Pendentes_28jun24.csv")
path_relativo = ajuste_path(path)
```

Escolha de colunas do dataframe

```python
df = df[['Ordem', 'Local de instalação', 'Trabalho',
         'Duração normal', 'Tipo de atividade']]
```

Função para contar o número de pessoas envolvidas por operação

```python
df["Pessoas"] = np.nan


def conta_pessoas(row):
    '''
    Conta o numero de pessoas por operação em uma nova coluna "Pessoas".
    Esse cálculo é realizado pela divisão do trabalho pela duração.
    A duração indica a quantidade de HH praticada por pessoa em uma operação.
    '''
    if (row['Trabalho'] != 0) & (row['Duração normal'] != 0):
        return row['Trabalho'] / row['Duração normal']
    return 0


df['Pessoas'] = df.apply(conta_pessoas, axis=1)
df
```

Função para arredondar números quebrados 

```python
def arredonda(row):
    '''
    Função para arredondar números quebrados. 
    '''
    pessoas = row['Pessoas']
    if (pessoas % 1) >= 0.5:
        return (pessoas//1)+1
    return (pessoas//1)


df['Pessoas'] = df.apply(arredonda, axis=1)
```
