### Importando bibliotecas necessárias

```python
import pandas as pd
from resolve_path import ajuste_path, read_input
from unidecode import unidecode
```

### Leitura das planilhas

```python
path = "data/"
pathIn = path + "input/"
pathUtil = path + "util/"
pathUtil = ajuste_path(pathUtil)

iw49 = read_input("IW49_OP_Pendentes_28jun24.csv")
zpm = read_input("ZPM_DATAS_ORDEM_Pendentes-28jun24 1.csv")
```

### Seleção das colunas necessárias

```python
colunas_IW49 = [
    "Ordem",
    "Trabalho",
    "Duração normal",
    "Confirmação"
]

iw49_filtrada = iw49[colunas_IW49]
```

```python
colunas_ZPM = [
    "Nº ordem",
    "Local de instalação",
    "Tipo de Ordem",
    "Data de Referência",
]

zpm_filtrada = zpm[colunas_ZPM]
```

### Integração das duas tabelas por colunas 'Ordem' e 'Nº ordem'

```python
merged = pd.merge(iw49_filtrada, zpm_filtrada,
                  left_on="Ordem", right_on="Nº ordem")
merged = merged.dropna(subset=["Local de instalação"])
merged
```
