```python
import pandas as pd
from resolve_path import ajuste_path
```

```python
path = "data/util/"
path = ajuste_path(path)
df_acidentes = pd.read_csv(path + "acidentes/" + "acidentes_entrada.csv")
df_OS = pd.read_csv(path + "os/" + "os_entrada.csv")
print(df_acidentes.info())
print(df_OS.info())
```

```python
df_OS.head()
```

```python
df_acidentes.head()
```

```python
data_min = df_acidentes["Ano Mes"].min()
data_max = df_acidentes["Ano Mes"].max()
df_OS = df_OS.loc[(df_OS["Ano Mes"] >= data_min) &
                  (df_OS["Ano Mes"] <= data_max)]
print(df_OS.head(), "\n", df_OS.tail())
```

```python
dataset_treinamento = pd.merge(
    df_OS, df_acidentes, on=["Ano Mes", "Local de instalação"], how="left"
)
dataset_treinamento.fillna(0, inplace=True)
```

```python
dataset_treinamento['Ano'] = dataset_treinamento['Ano Mes'].str[:4].astype(int)
dataset_treinamento['Mes'] = dataset_treinamento['Ano Mes'].str[5:7].astype(
    int)
dataset_treinamento.drop(columns=['Ano Mes'], axis=1, inplace=True)

dataset_treinamento.columns
```

#### Ajustando a ordem das colunas na tabela de saída

```python
cols = [
    "Ano",
    "Mes",
    "Local de instalação"
]

cols_acidente = df_acidentes.columns.to_list()
cols_OS = df_OS.columns.to_list()

cols_acidente.remove("Ano Mes")
cols_OS.remove("Ano Mes")

for c in cols:
    if c in ["Ano", "Mes"]:
        continue
    cols_acidente.remove(c)
    cols_OS.remove(c)

cols.extend(cols_acidente)
cols.extend(cols_OS)

dataset_treinamento = dataset_treinamento[cols]
```

```python
dataset_treinamento.to_csv(path + "dataset_treinamento.csv", encoding='utf-8')
```
