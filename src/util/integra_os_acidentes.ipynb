```python
import pandas as pd
```

```python
path = "../../data/util/"

df_acidentes = pd.read_csv(path + "acidentes/" + "acidentes_entrada.csv")
df_OS = pd.read_csv(path + "os/" + "os_entrada.csv")
print(df_acidentes.info())
print(df_OS.info())
```

```python
df_OS.head()
```

```python
df_acidentes.head()
```

```python
data_min = df_acidentes["Ano Mes"].min()
data_max = df_acidentes["Ano Mes"].max()
df_OS = df_OS.loc[(df_OS["Ano Mes"] >= data_min) &
                  (df_OS["Ano Mes"] <= data_max)]
print(df_OS.head(), "\n", df_OS.tail())
```

```python
dataset_entrada = pd.merge(
    df_OS, df_acidentes, on=["Ano Mes", "Local de instalação"], how="left"
)
dataset_entrada.fillna(0, inplace=True)
```

```python
dataset_entrada['Ano'] = dataset_entrada['Ano Mes'].str[:4].astype(int)
dataset_entrada['Mes'] = dataset_entrada['Ano Mes'].str[5:7].astype(int)
dataset_entrada.drop(columns=['Ano Mes'], axis=1, inplace=True)
```

```python
dataset_entrada = dataset_entrada[
    [
        "Ano",
        "Mes",
        "Local de instalação",
        "HH",
        "Quantidade de Acidentes",
        "Quantidade de Acidentes de Alto Potencial",
        "Binario Acidentes de Alto Potencial",
    ]
]
```

```python
dataset_entrada.to_csv(path + "dataset_entrada.csv", encoding='utf-8')
```
